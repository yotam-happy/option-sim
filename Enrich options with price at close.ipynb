# Fill in missing days for stock prices

```python
# load pre calculated augmented stock data prices file, instead of calculating again
#prices_df = pd.read_csv('E:/projects/stock data/only_stock_prices_filled_in.csv', parse_dates=['DataDate'])
```

```python
# load stock data prices file
prices_df = pd.read_csv('E:/projects/stock data/only_stock_prices.csv', parse_dates=['DataDate'])
```

```python
# should already be sorted, but make sure
prices_df = prices_df.sort_values(['Symbol', 'DataDate'])
```

```python
# maximum days to fill in missing values
max_fillin_days = 5
```

```python
rows = []
for i, (ticker, ticker_df) in enumerate(prices_df.groupby('Symbol')):
    print(i, ticker)
    last_row = None
    for row in ticker_df.itertuples():
        row = row._asdict()
        del row['Index']
        if last_row is not None and (row['DataDate'] - last_row['DataDate']).days > 1:
            for i in range(1, min((row['DataDate'] - last_row['DataDate']).days, max_fillin_days)):
                r = last_row.copy()
                r['DataDate'] = r['DataDate'] + timedelta(days=i)
                rows.append(r)
        rows.append(row)
        last_row = row
```

```python
# save augmented stock prices
prices_df = pd.DataFrame(rows)
```

```python
prices_df.to_csv('E:/projects/stock data/only_stock_prices_filled_in.csv', index=False)
```
